### Q5. Check for the Null values in the dataframe

```python
# Check for null values in the DataFrame
df.isnull().sum()
```

### Q6. Replace all the null values with "NaN"

```python
# Replace all null values with NaN
df.replace(to_replace=[None, ''], value=np.nan, inplace=True)
```

### Q7. Change the format of date columns - "ORDER_CREATION_DATE" to datetime[64] with the format as "%Y%m%d"

```python
# Change the format of ORDER_CREATION_DATE to datetime[64] with format "%Y%m%d"
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format='%Y%m%d')
```

### Q8. Do the same activity for the other date field i.e. "REQUESTED_DELIVERY_DATE" to datetime[64] with the format as "%Y%m%d"

```python
# Change the format of REQUESTED_DELIVERY_DATE to datetime[64] with format "%Y%m%d"
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')
```

### Q9. Sanity check - Check how many records are having order date greater than the delivery date

```python
# Count how many records have order date greater than delivery date
print((df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']).sum())
```

### Q10. Remove those records where order date is greater than the delivery date

```python
# Remove records where order date is greater than delivery date
df = df[df['ORDER_CREATION_DATE'] <= df['REQUESTED_DELIVERY_DATE']]
```

### Q11. Check the number of records where the “ORDER_AMOUNT” field is having “-” in it..

```python
# Count records where "ORDER_AMOUNT" field has "-"
print(df[df["ORDER_AMOUNT"] == "-"].shape[0])
```

### Q12. Replace “-” with “” from the “ORDER_AMOUNT” field. 

```python
# Replace "-" with empty string in "ORDER_AMOUNT" field
df["ORDER_AMOUNT"] = df["ORDER_AMOUNT"].replace("-", "")
```

### Q13. Check the number of records where the “ORDER_AMOUNT” field is having “,” in it..

```python
# Count the number of records with ","
comma_count = df['ORDER_AMOUNT'].str.contains(",", na=False).sum()
comma_count
```

### Q14. Replace “,” with “.” from the “ORDER_AMOUNT” field.

```python
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace(",", ".")
```

### Q15. Count the number of records where the order date and the delivery date are same

```python
# Count the number of records where order date and delivery date are the same
same_date_count = (df['ORDER_CREATION_DATE'] == df['REQUESTED_DELIVERY_DATE']).sum()
same_date_count
```

### Q16. Count the number of records for each currency type by using the field “'ORDER_CURRENCY'”

```python
# Count the number of records for each currency type
currency_counts = df['ORDER_CURRENCY'].value_counts()
currency_counts
```

### Q17. Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 

```python
# Create a new column 'amount_in_usd' and convert non-USD currencies to USD
conversion_rate = {
    'USD': 1.0,
    'EUR': 1.18,
    'AUD': 0.75,
    'CAD': 0.80,
    'GBP': 1.37,
    'MYR': 0.24,
    'PLN': 0.26,
    'AED': 0.27,
    'HKD': 0.13,
    'CHF': 1.09,
    'RON': 0.24,
    'SGD': 0.74,
    'CZK': 0.045,
    'HU1': 0.0036,
    'NZD': 0.70,
    'BHD': 2.65,
    'SAR': 0.27,
    'QAR': 0.27,
    'KWD': 3.32,
    'SEK': 0.11
}
df['ORDER_AMOUNT'] = pd.to_numeric(df['ORDER_AMOUNT'], errors='coerce')    #converted string values to numeric
# df['AMOUNT_IN_USD'] = df.apply(lambda row: row['ORDER_AMOUNT'] * usd_conversion_rates.get(row['ORDER_CURRENCY'], np.nan), axis=1)
df['AMOUNT_IN_USD'] = df.apply(lambda row: row['ORDER_AMOUNT'] * conversion_rate.get(row['ORDER_CURRENCY'], 0) if row['ORDER_AMOUNT'] != 0 else 0, axis=1)
df
```

### Q18. Check for values “0” in the “'amount_in_usd” column. 

```python
# Check for values "0" in the 'amount_in_usd' column
zero_amount_count = (df['AMOUNT_IN_USD'] == 0).sum()
zero_amount_count
```

### Q19.  Create a new column in the existing dataframe “unique_cust_id” by adding 'CUSTOMER_NUMBER' and 'COMPANY_CODE'

```python
# Create a new column 'unique_cust_id' by combining 'CUSTOMER_NUMBER' and 'COMPANY_CODE'
df['UNIQUE_CUST_ID'] = df['CUSTOMER_NUMBER'].astype(str) + df['COMPANY_CODE'].astype(str)
df
```
