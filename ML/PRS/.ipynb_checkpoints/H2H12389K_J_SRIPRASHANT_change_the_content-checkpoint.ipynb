Create a new column in the DataFrame called "amount_in_usd" and convert all non-USD currencies to USD

Creating a Dictionary that has all the currencies's conversion rate

```python
usd_conversion_rates = {
    'USD': 1.0,
    'EUR': 1.07,
    'AUD': 0.65,
    'CAD': 0.74,
    'GBP': 1.24,
    'MYR': 0.22,
    'PLN': 0.24,
    'AED': 0.27,
    'HKD': 0.13,
    'CHF': 1.10,
    'RON': 0.22,
    'SGD': 0.74,
    'CZK': 0.045,
    'HU1': 0.0029,
    'NZD': 0.61,
    'BHD': 2.65,
    'SAR': 0.27,
    'QAR': 0.27,
    'KWD': 3.25,
    'SEK': 0.093
}
```


converting ORDER_AMOUNT and RELEASED_CREDIT_VALUE coloumn to numeric and converting every amount in USD
- using the Dictionary that we created we are creating a new coloumn that has order amount in USD only 
- we created a Lambda function that takes the row of the dataframe as input and multiplies the amount in ORDER_AMOUNT and RELEASED_CREDIT_VALUE coloumn resplectively  with the conversion rates present in the dictionary which is indexed by ORDER_CURRENCY
- if the currency is not present in the dictionary it returns NaN instead
- axis=1 specifies that lambda function must be applied to each row   

```python

data['AMOUNT_IN_USD'] = data.apply(lambda row: row['ORDER_AMOUNT'] * usd_conversion_rates.get(row['ORDER_CURRENCY'], np.nan), axis=1)
data['RELEASED_CREDIT_VALUE'] = data.apply(lambda row: row['RELEASED_CREDIT_VALUE'] * usd_conversion_rates.get(row['ORDER_CURRENCY'], np.nan), axis=1)
```

This barplot shows the frequency distribution of different distribution channels in a dataset.
- The median is used as a threshold here to filter out the values that occur less frequently than a certain cutoff point. In this case, the median value is calculated using the statistics.median() function on the frequency counts of the distribution channels.
- The idea behind using the median as a threshold is to only show the most commonly occurring distribution channels, thereby simplifying the visualization and making it easier to interpret.
- Value_counts() is useful for quickly understanding the distribution of values in a categorical variable.

Why Do We need this barplot?
-  This barplot provides a visual summary of the distribution of distribution channels (countries) in the dataset, highlighting the most frequently occurring ones. This can be useful in identifying potential patterns or biases in the data related to sales by country.

How it is better than the Histogram?
- A histogram is typically used to display the distribution of continuous numerical data, where the data is divided into intervals or bins and plotted as bars.
- In contrast, a barplot shows the count or frequency of each category as separate bars, which makes it suitable for displaying categorical data.

WE CAN SEE THAT KOREA CHANNEL IS WILDLY USED FOLLOWED BY CONGO

```python
#calculation of the median using statistics library
Median = statistics.median(data['DISTRIBUTION_CHANNEL'].value_counts())

# Count the frequency of each value in the DISTRIBUTION_CHANNEL column
counts = data['DISTRIBUTION_CHANNEL'].value_counts()

# Filter out the values that occur less frequently than a certain threshold
# Here the threshold is median
threshold = Median
counts = counts[counts >= threshold]

# Plot a horizontal bar chart to show the top N values by frequency
plt.figure(figsize=(20, 18))
sns.barplot(x=counts.values, y=counts.index)
plt.xlabel('Frequency')
plt.ylabel('Distribution Channel')
plt.title(f'Top {len(counts)} Distribution Channels')
plt.show()
```

why did we sort and group and filter the data?
- The purpose of sorting the data based on AMOUNT_IN_USD is to arrange the data in descending order of the purchase amount, so that the top values can be identified and analyzed.
- Grouping the data by PURCHASE_ORDER_TYPE and DISTRIBUTION_CHANNEL allows for an aggregation of all orders that fall under each category. This helps to identify the number of orders for each purchase order type and distribution channel.
- Filtering the dataframe to include only the top 20 PURCHASE_ORDER_TYPE values and top 20 DISTRIBUTION_CHANNEL.
- The cross-tabulation created using pd.crosstab() helps to identify the relationship between the PURCHASE_ORDER_TYPE and DISTRIBUTION_CHANNEL variables. The stacked bar chart built from this cross-tabulation allows us to visualize the count of orders in each category.

All this is done to narrow down the data and focus on the most significant details, which can then be used to create insights.

```python
# Sorting the data based on AMOUNT_IN_USD 
data_sorted = data.sort_values(by='AMOUNT_IN_USD', ascending=False)

# Get the top 20 values for PURCHASE_ORDER_TYPE and DISTRIBUTION_CHANNEL
top_purchase_orders = data_sorted['PURCHASE_ORDER_TYPE'].value_counts().head(20).index.tolist()
top_dist_channels = data_sorted['DISTRIBUTION_CHANNEL'].value_counts().head(20).index.tolist()

# Filter the dataframe to include only the top 20 PURCHASE_ORDER_TYPE values and top 20 DISTRIBUTION_CHANNEL values
data_filtered = data[(data['PURCHASE_ORDER_TYPE'].isin(top_purchase_orders)) & (data['DISTRIBUTION_CHANNEL'].isin(top_dist_channels))]

# Create a cross-tabulation of PURCHASE_ORDER_TYPE and DISTRIBUTION_CHANNEL
ct = pd.crosstab(data_filtered['PURCHASE_ORDER_TYPE'], data_filtered['DISTRIBUTION_CHANNEL'])

# Create a stacked bar chart of the cross-tabulation
fig, ax = plt.subplots(figsize=(12,8))
ct.plot(kind='bar', stacked=True, ax=ax)

plt.title('Top 20 Purchase Order Types and Distribution Channels by AMOUNT_IN_USD')
plt.ylabel('Number of Orders')
plt.xlabel('Purchase Order Type')
plt.legend(loc='best')

plt.show()
```

REPLACING OUTLIERS FROM THE COLOUMN AMOUNT_IN_USD WITH THE MEDIAN

WHY MEDIAN?
 - We could not use Mode as the number of zeros's was in highest freaquncy
 - we could not use mean as it is sensitive to extreme values
 - Median should best represent the data

WHY IQR METHOD TO FIND OUT THE OUTLIERS?
 - IQR is more robust in outliers ans less affected by extreme values
 - The IQR is distribution-agnostic and can be applied to any type of distribution.On the other hand zscore is applied only on normal distribution
 

CHANGES MADE IN MILESTONE 4
- previously used median to replace the outliers
- now, we use Q3 value only to remove the outliers on the higher end

WHY Q3 value?
- As it is a Time series data replacing by the by median we loose the speciality of the time series data or our ML algos cannot caputre the trend

```python
# Calculate the median of the "AMOUNT_IN_USD" column
# amount_median = data['AMOUNT_IN_USD'].median()

# Calculate the IQR of the "AMOUNT_IN_USD" column
q1 = data['AMOUNT_IN_USD'].quantile(0.25)
q3 = data['AMOUNT_IN_USD'].quantile(0.75)
iqr = q3 - q1
#replace the outliers by q3
replacing_ouliers = q3
# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Replace outliers with the mode value
data['AMOUNT_IN_USD'] = data['AMOUNT_IN_USD'].apply(lambda x: replacing_ouliers if x < lower_bound or x > upper_bound else x)
```

- The lagged sales values provide information about the sales performance of each customer in the previous days.
- These lagged sales and differences can be used as features in a predictive model to understand how past sales patterns relate to the 'AMOUNT_IN_USD' and potentially improve predictions.

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
data = difference_in_days(data, lags=5, ffday=1, customer_id_col='UNIQUE_CUST_ID', create_date_col='ORDER_CREATION_DATE', net_amount_col='AMOUNT_IN_USD')
```

```python
data['Mean_Amount_by_SalesOrg_Division'] = data.groupby(['SALES_ORG', 'DIVISION'])['AMOUNT_IN_USD'].transform('mean')
```

The groupby(['SALES_ORG', 'DIVISION']) part groups the data based on unique combinations of 'SALES_ORG' and 'DIVISION
.Transform('mean') function is applied to the 'AMOUNT_IN_USD' column within each group. It calculates the mean value of 'AMOUNT_IN_USD' for each group.

WHY DO WE NEED THIS?
- The 'Mean_Amount_by_SalesOrg_Division' column provides insights into the average order amount in USD for each unique combination of sales organization and division. It helps us understand the spending patterns or purchasing behavior specific to different sales organizations and divisions.
- By analyzing this column, we can identify sales organizations and divisions that tend to generate higher or lower average order amounts.

BEFORE SPLITTING THE DATA WE SORT IT ON THE MONTH AND THE DAYS IN IT.

why?
- As it is a time series data so to capture the pattern based on time we have sorted the data

```python
sorted_data = data.sort_values(by=['ORDER_CREATION_MONTH', 'ORDER_CREATION_DAY'])
```

Using the train_test split of the sklearn library to split the data into train and test
- shuffle = False is done so that the while splitting the sampling is not done by train test split as we sorted the data on time

```python
from sklearn.model_selection import train_test_split
# Split the data into features and target variable
X = sorted_data.drop(['AMOUNT_IN_USD', 'ORDER_CREATION_MONTH', 'ORDER_CREATION_DAY','CUSTOMER_ORDER_ID','CREDIT_CONTROL_AREA','DAYS_LEFT'], axis=1)

y = sorted_data['AMOUNT_IN_USD']
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
#following 70:30 ratio of training and test sets
```

IMPLEMENTING LINEAR REGRESSION
- USING THE LinearRegression module fro sklearn library

```python
#initializing dictionaries to capture the R square value of every model that we have Implemented 
model_names = []
r_squared_values = []
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Initialize and train the regression models
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
# Make predictions on the test data
linear_regression_pred = linear_regression.predict(X_test)
# Evaluate the models
lr_mse = mean_squared_error(y_test, linear_regression_pred)
lr_rmse = mean_squared_error(y_test, linear_regression_pred, squared=False)
lr_mae = mean_absolute_error(y_test, linear_regression_pred)
lr_r2 = r2_score(y_test, linear_regression_pred)
model_names.append('LINEAR REG')
r_squared_values.append(lr_r2)

print("Linear Regression:")
print("MSE:",lr_mse)
print("RMSE:", lr_rmse)
print("MAE:", lr_mae)
print("R-Squared:",lr_r2)

```

IMPLEMENTING DESCISION TREES

```python
from sklearn.tree import DecisionTreeRegressor
# Initialize and train the  models
decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train, y_train)
# Make predictions on the test data
decision_tree_pred = decision_tree.predict(X_test)
# Evaluate the models
dt_mse = mean_squared_error(y_test, decision_tree_pred)
dt_rmse = mean_squared_error(y_test, decision_tree_pred, squared=False)
dt_mae = mean_absolute_error(y_test, decision_tree_pred)
dt_r2 = r2_score(y_test, decision_tree_pred)
model_names.append('DESCISION TREE')
r_squared_values.append(dt_r2)

print("Decision Tree:")
print("MSE:", dt_mse)
print("RMSE:", dt_rmse)
print("MAE:", dt_mae)
print("R-Squared:", dt_r2)
print()
```

# HYPERPARAMETER TUNING
- WE MAKE THE TUNING OF THE XGBOOST AS IT IS SHOWING GREATER ACCURACY AMONG OTHER ML MODELS

- We have selected RANDOMIZED SEARCH CV as the data is very huge and Gridsearch cv would have taken lot of time, also our machines would not support that much load

- AFTER TUNING R-Squared VALUE COMES OUT TO BE 50%,  WHICH IS RELEVANT FOR A TIME SERIES DATA

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid for randomized search
xgb_params = {
    'booster': ['gbtree'],
    'objective': ['reg:squarederror'],
    'eta': [0.1, 0.3, 0.5],
    'gamma': [0, 0.1, 0.2],
    'max_depth': [4, 6, 8],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'lambda': [0.1, 1.0, 10.0],
    'alpha': [0, 0.1, 0.5],
    'tree_method': ['auto', 'exact', 'approx'],
    'n_jobs': [-1],
    'random_state': [0]
}


# Create the RandomizedSearchCV object
xgb_random = RandomizedSearchCV(estimator=XGBRegressor(), param_distributions=xgb_params, scoring='neg_mean_squared_error', n_iter=10, cv=5, random_state=42)

# Fit the randomized search on the training data
xgb_random.fit(X_train, y_train)

# Get the best estimator and make predictions on the test data
xgb_best = xgb_random.best_estimator_
xgb_pred_best = xgb_best.predict(X_test)

# Calculate evaluation metrics
xgb_mse_best = mean_squared_error(y_test, xgb_pred_best)
xgb_rmse_best = mean_squared_error(y_test, xgb_pred_best, squared=False)
xgb_mae_best = mean_absolute_error(y_test, xgb_pred_best)
xgb_r2_best = r2_score(y_test, xgb_pred_best)

# Print the evaluation metrics
print("XGBoost (Best Parameters):")
print("MSE:", xgb_mse_best)
print("RMSE:", xgb_rmse_best)
print("MAE:", xgb_mae_best)
print("R-Squared:", xgb_r2_best)
```
